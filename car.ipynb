# Importamos Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Creamos el dataframe

```python
df = pd.read_csv("C:/Users/maxir/OneDrive/Escritorio/prueba/dataset.csv", delimiter=",")
df.info()
```

# Eliminación de columnas innecesarias

```python
df.drop(["Car_id"] ,axis=1).drop(["Gender"] ,axis=1).drop(["Model"] ,axis=1).drop(["Transmission"] ,axis=1).drop(["Customer Name"] ,axis=1).drop(["Dealer_Name"] ,axis=1).drop(["Phone"] ,axis=1).drop(["Color"] ,axis=1).drop(["Annual Income"], axis=1).drop(["Price ($)"], axis=1).drop(["Body Style"], axis=1).drop(["Dealer_Region"], axis=1).drop(["Engine"], axis=1).drop(["Dealer_No "], axis=1)
```

```python
df["Company"]
```

```python
chevrolet = df[df["Company"] == "Chevrolet"]
ford = df[df["Company"] == "Ford"]
toyota = df[df["Company"] == "Toyota"]
honda = df[df["Company"] == "Honda"]
volkswagen = df[df["Company"] == "Volkswagen"]
mercedez_benz = df[df["Company"] == "Mercedes-B"]
audi = df[df["Company"] == "Audi"]
nissan = df[df["Company"] == "Nissan"]
bmw = df[df["Company"] == "BMW"]
hyundai = df[df["Company"] == "Hyundai"]

cars = [len(chevrolet), len(ford), len(toyota), len(honda), len(volkswagen), len(mercedez_benz), len(audi), len(nissan), len(bmw), len(hyundai)]
nombres = ["chevrolet", "ford", "Toyota", "Honda", "Volkswagen", "Mercedez Benz", "Audi", "Nissan", "BMW", "Hyudai"]
```

# Graficar los resultados

```python
fig, ax = plt.subplots()

# Definimos el tamaño
plt.figure(1, figsize=(12))

# Creamos la torta
ax.pie(cars, labels=nombres, autopct='%1.1f%%', shadow=True, startangle=110)

# Titulo
plt.title("Cantidad")

plt.show()
```
